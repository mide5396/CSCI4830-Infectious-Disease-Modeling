```python
%matplotlib ipympl
import matplotlib.pyplot as plt
import numpy as np
```

```python
def ForwardEulerSolver(N, I_initial, S_initial, Beta, Gamma):
    
    # Set x limit
    xRange = 50
    
    
    dt = 0.01
    
    # Find initial R
    # In this problem, the initial R would be 0
    R_initial = N - (S_initial + I_initial)
    
    S = [S_initial]
    I = [I_initial]
    R = [R_initial]
    
    t = [0]
    
    # Find number of steps until you reach xRange and loop through them
    for n in range(int(xRange / dt)):
        
        dS = - Beta * S[n] * I[n] / N
        S_nPlusOne = S[n] + dt * dS
        S.append(S_nPlusOne)
        
        dI = Beta * S[n] * I[n] / N - Gamma * I[n]
        I_nPlusOne = I[n] + dt * dI
        I.append(I_nPlusOne)
        
        dR = Gamma * I[n]
        R_nPlusOne = R[n] + dt * dR
        R.append(R_nPlusOne)
        
        t_nPlusOne = t[n] + dt
        t.append(t_nPlusOne)
    return [t,S,I,R]
```

```python
fig, axs = plt.subplots(3, figsize=(8, 15))

N = 1000
I_initial = 1
S_initial = 999

#Beta = 1, Gamma = 0.5
t0,S0,I0,R0 = ForwardEulerSolver(N, I_initial, S_initial, 1, 0.5)
axs[0].cla()
axs[0].plot(t0, S0, label='S Michael')
axs[0].plot(t0, I0, label='I Michael')
axs[0].plot(t0, R0, label='R Michael')
axs[0].set_xlabel("Time")
axs[0].set_ylabel("People")
axs[0].legend()
axs[0].set_title("SIR for N = 1000, $I_0$ = 1, $S_0$ = 999, $\u03B2$ = 1, $\u03B3$ = 0.5")

#Beta = 1.5, Gamma = 0.5
t1,S1,I1,R1 = ForwardEulerSolver(N, I_initial, S_initial, 1.5, 0.5)
axs[1].cla()
axs[1].plot(t1, S1, label='S Michael')
axs[1].plot(t1, I1, label='I Michael')
axs[1].plot(t1, R1, label='R Michael')
axs[1].set_xlabel("Time")
axs[1].set_ylabel("People")
axs[1].legend()
axs[1].set_title("SIR for N = 1000, $I_0$ = 1, $S_0$ = 999, $\u03B2$ = 1.5, $\u03B3$ = 0.5")

# Beta = 2, Gamma = 0.5
t2,S2,I2,R2 = ForwardEulerSolver(N, I_initial, S_initial, 2, 0.5)
axs[2].cla()
axs[2].plot(t2, S2, label='S Michael')
axs[2].plot(t2, I2, label='I Michael')
axs[2].plot(t2, R2, label='R Michael')
axs[2].set_xlabel("Time")
axs[2].set_ylabel("People")
axs[2].legend()
axs[2].set_title("SIR for N = 1000, $I_0$ = 1, $S_0$ = 999, $\u03B2$ = 2, $\u03B3$ = 0.5")
plt.show()

plt.savefig('problem1Graphs.png')
```
