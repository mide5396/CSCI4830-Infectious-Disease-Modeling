```python
%matplotlib ipympl
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import fsolve
```

```python
N = 10**6
epsilon = 1 / N
steps = 50
S = np.linspace(1, N, steps)
Beta = 1
Gamma = .5

one_over_R_naught = Gamma / Beta

episilon_deriv = (Beta * (S / N) - Gamma) * epsilon
plt.cla()
plt.plot(S / N, episilon_deriv)

plt.axvline(x = one_over_R_naught, color = 'r', label = '1 / $R_0$', linestyle='dashed') 

plt.legend()

plt.xlabel("s")
plt.ylabel("$\epsilon\u0307$")
plt.title("$\epsilon\u0307$ as a function of s, $\u03B2$ = 1, $\u03B3$ = 0.5")
plt.show()
plt.savefig('problem4Graph.png')
```
